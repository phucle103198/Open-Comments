### Delete number, ":" is not in Vietnamese

```python
def remove_non_vietnamese(sentiment_data):
    if isinstance(sentiment_data, str) or isinstance(sentiment_data, int) or isinstance(sentiment_data, float):
        if isinstance(sentiment_data, str):
            sentiment_data = re.sub(r'\d+', '', sentiment_data)
            sentiment_data = re.sub(r':', '', sentiment_data)
            return re.sub(r'[^\wÀ-ỹ\s.,/]', ' ', sentiment_data)
        else:
            return str(sentiment_data)
    else:
        return np.nan

data_df.replace('\n', '. ', regex=True, inplace=True)

data_df = data_df.applymap(remove_non_vietnamese)
data_df.head()
```

```python
def tach_tu(data):
    text = 'text'
    upos = 'upos'
    head = 'head'
    deprel = 'deprel'
    feats = 'feats'
    temp = []

    for item in data:
        for token in item['tokens']:
            temp.append({
                text: token[text],
                upos: token[upos],
                head: token[head],
                deprel: token[deprel],
                feats: token[feats].split('=')[1] if token.get(feats) is not None else None
            })
        temp.append({
            text: None,
            upos: None,
            head: None,
            deprel: None,
            feats: None
        })

    return temp
```

## Remove words from dictionary

```python
def flat_array(input_arr):
    flat_arr = []
    flat_id_counter = 1
    text = 'text'
    upos = 'upos'
    head = 'head'
    deprel = 'deprel'
    feats = 'feats'

    for sublist in input_arr:
        if sublist == None:
            flat_arr.append(None)
            continue
        flat_sublist = []
        for item in sublist:
            if ' ' in item[text]:
                words = item[text].split()
                for _, word in enumerate(words):
                    flat_sublist.append({
                        text: word,
                        upos: item[upos],
                        head: item[head],
                        deprel: item[deprel],
                        feats: item[feats],
                        'flat_id': flat_id_counter
                    })
                flat_id_counter += 1
            else:
                flat_sublist.append({
                    text: item[text],
                    upos: item[upos],
                    head: item[head],
                    deprel: item[deprel],
                    feats: item[feats]
                })
        flat_arr.append(flat_sublist)

    return flat_arr

def unflat_array(flat_arr):
    text = 'text'
    upos = 'upos'
    head = 'head'
    deprel = 'deprel'
    feats = 'feats'
    my_arr = []
    for sublist in flat_arr:
        if sublist == None:
            my_arr.append(None)
            continue
        unflat_sublist = []
        uncombined_arr = []
        for item in sublist:
            if 'flat_id' in item:
                if len(uncombined_arr) > 0 and item['flat_id'] != uncombined_arr[0]['flat_id']:
                    combined_text = ' '.join(item[text] for item in uncombined_arr)
                    combined_item = {
                        text: combined_text,
                        upos: uncombined_arr[0][upos],
                        head: uncombined_arr[0][head],
                        deprel: uncombined_arr[0][deprel],
                        feats: uncombined_arr[0][feats]
                    }
                    unflat_sublist.append(combined_item)
                    uncombined_arr = []
                uncombined_arr.append(item)
            else:
                if len(uncombined_arr) > 0:
                    combined_text = ' '.join(item[text] for item in uncombined_arr)
                    combined_item = {
                        text: combined_text,
                        upos: uncombined_arr[0][upos],
                        head: uncombined_arr[0][head],
                        deprel: uncombined_arr[0][deprel],
                        feats: uncombined_arr[0][feats]
                    }
                    unflat_sublist.append(combined_item)
                    uncombined_arr = []
                unflat_sublist.append(item)
        if len(uncombined_arr) > 0:
            combined_text = ' '.join(item[text] for item in uncombined_arr)
            combined_item = {
                text: combined_text,
                upos: uncombined_arr[0][upos],
                head: uncombined_arr[0][head],
                deprel: uncombined_arr[0][deprel],
                feats: uncombined_arr[0][feats]
            }
            unflat_sublist.append(combined_item)
        my_arr.append(unflat_sublist)

    return my_arr
```

```python
tu_can_loai_df = pd.read_excel('./Open Comments/tu_can_loai.xlsx')
tu_can_loai = tu_can_loai_df['Từ Cần Loại'].tolist()
tu_can_loai = sorted(tu_can_loai, key=len, reverse=True)

def loai_tu(data):
    data = copy.deepcopy(data)
    text = 'text'
    data = flat_array(data)
    lenData = len(data)
    for tuCL in tu_can_loai:
        i = 0
        tu_arr = tuCL.split(' ')
        for i in range(lenData):
            if data[i] == None:
                continue
            j = 0
            lenCau = len(data[i])
            while j < lenCau:
                if data[i][j][text].lower() == tu_arr[0]:
                    isMatch = True
                    tuCount = len(tu_arr)
                    if j + tuCount <= lenCau:
                        for k in range(1, tuCount):
                            if data[i][j + k][text].lower() != tu_arr[k]:
                                isMatch = False
                                break
                    else:
                        isMatch = False
                    if isMatch:
                        for _ in range(tuCount):
                            del data[i][j]
                        lenCau -= tuCount
                        j -= 1
                j += 1
    data = unflat_array(data)
    return data
```

## Sentences Separation

```python
def tach_cau(data):
    text = 'text'
    upos = 'upos'
    head = 'head'
    deprel = 'deprel'
    feats = 'feats'

    tach_cau_data = []
    cau = []
    stop_point = len(data)
    i = 0

    while i < stop_point:
        if data[i][text] is None:
            if (len(cau) > 0):
                tach_cau_data.append(cau)
                cau = []
            tach_cau_data.append(None)
            i += 1
            continue
        if len(data[i][text].split(' ')) > 1 and data[i][text].split(' ')[0] == data[i][text].split(' ')[1]:
            data[i][text] = data[i][text].split(' ')[0]
        if data[i][upos] is None:
            tach_cau_data.append(cau)
            cau = []
            i += 1
            continue
        if data[i][upos] == 'NOUN':
            cau.append(data[i])
            i += 1
            continue
        if data[i][upos] == 'VERB':
            if data[i][deprel] == 'aux:pass':
                i += 1
                continue
            elif len(cau) > 0:
                cau.append(data[i])
                i += 1
                continue
            else:
                i += 1
                continue
        if data[i][upos] == 'X':
            if data[i][feats] == 'Neg':
                cau.append(data[i])
                i += 1
                continue
            else:
                if data[i + 1][upos] == 'ADJ':
                    if data[i + 2][upos] == 'X':
                        cau.append(data[i + 1])
                        cau.append(data[i + 2])
                        i += 3
                        continue
                    else:
                        cau.append(data[i])
                        cau.append(data[i + 1])
                        i += 2
                        continue
                i += 1
                continue
        if data[i][upos] == 'SCONJ':
            tach_cau_data.append(cau)
            if data[i + 1][upos] == 'NOUN':
                cau = [data[i + 1]]
                i += 2
                continue
            else:
                cau = [x for x in cau if x[upos] == 'NOUN'][:1]
                i += 1
                continue
        if data[i][upos] == 'ADP':
            i += 1
            continue
        if data[i][upos] == 'CCONJ':
            i += 1
            continue
        if data[i][upos] == 'PROPN':
            i += 1
            continue
        if data[i][upos] == 'ADJ':
            if data[i + 1][upos] == 'X':
                cau.append(data[i])
                cau.append(data[i + 1])
                i += 2
                continue
            cau.append(data[i])
            i += 1
            continue
        if data[i][upos] == 'PUNCT':
            tach_cau_data.append(cau)
            cau = []
            i += 1
            continue

        i += 1

    stop_point = len(tach_cau_data)
    i = 0

    while i < stop_point:
        if tach_cau_data[i] is not None and len(tach_cau_data[i]) == 1 and tach_cau_data[i][0][upos] != 'ADJ':
            tach_cau_data[i] = [{text: 'Loai bo'}]
        i += 1

    return [x for x in tach_cau_data if x is None or (len(x) > 0 and x[0][text] != 'Loai bo')]
```

```python
def arr2sentences(data, name):
    text = 'text'
    sentences = []
    for sentence in data:
        if sentence is None:
            sentences.append({name: None})
            continue
        wordList = []
        for tu in sentence:
            wordList.append(tu[text])
        if (len(wordList) > 0):
            t = ' '.join(wordList)
            if (len(t) > 0):
                sentences.append({name: t})

    return sentences
```

## Count Frequency of Attribute

```python
strip_and_lower = lambda x: x.strip().lower() if isinstance(x, str) else x

def count_freq(data, keywords):
    sentences_data = copy.deepcopy(data)
    keywordList = []
    for kls in keywords:
        kl = pd.Series(kls.split(',')).apply(strip_and_lower).tolist()
        for i in range(len(kl)):
            keywordList.append({'key': kl[0], 'value': kl[i]})
    keywordList = sorted(keywordList, key=lambda x: len(x['value']), reverse=True)
    freq = {}
    for k in sentences_data.keys():
        for keyword in keywordList:
            if keyword['key'] not in freq:
                freq[keyword['key']] = { k: 0 }
            elif k not in freq[keyword['key']]:
                freq[keyword['key']][k] = 0
            for i in range(len(sentences_data[k])):
                if keyword['value'] in sentences_data[k][i]:
                    freq[keyword['key']][k] += 1
                    sentences_data[k][i] = sentences_data[k][i].replace(keyword['value'], '')
    return freq
```

```python
def match_sequence(sequence, search):
    upos = 'upos'
    if len(sequence) != len(search):
        return False
    for i in range(len(sequence)):
        if sequence[i][upos] != search[i]:
            return False
    return True

def removeWord(data, tk, luong):
    data = copy.deepcopy(data)
    for i in range(len(data)):
        for word in luong:
            if word in [x.lower() for x in data[i].split(' ')]:
                data[i] = data[i].replace(word, '').strip()
        for word in tk:
            if word == data[i].lower():
                data[i] = ''
    return data

def getOtherKeywords(data, keywordList, tu_dien_luong):
    SEARCH = [
        ['NOUN', 'NOUN', 'NOUN'],
        ['NOUN', 'NOUN'],
        ['NOUN', 'ADJ'],
        ['ADJ'],
    ]
    text = 'text'
    matched_phrases = []
    data = copy.deepcopy(data)

    for i in range(len(SEARCH)):
        j = 0
        while j < len(data):
            if data[j] is not None:
                k = 0
                while k < len(data[j]):
                    if k + len(SEARCH[i]) <= len(data[j]) and match_sequence(data[j][k:k+len(SEARCH[i])], SEARCH[i]):
                        matched_phrase = ' '.join(
                            [word[text] for word in data[j][k:k+len(SEARCH[i])]])
                        matched_phrases.append(matched_phrase)
                        for _ in range(len(SEARCH[i])):
                            del data[j][k]
                        k -= 1
                    k += 1
            j += 1

    raw = list(set(pd.Series(matched_phrases).apply(strip_and_lower)))
    removed = list(set(pd.Series(removeWord(matched_phrases, keywordList, tu_dien_luong)).apply(strip_and_lower)))
    return removed
```

```python
def format_data_by_people(data, name, point):
    result = []
    temp = {'sentences': []}
    for d in data:
        if d[name] is None:
            result.append(temp)
            temp = {'sentences': []}
        else:
            temp['sentences'].append(d[name])
            temp['liking'] = d[point]
    return result
```

```python
def phan_loai_luong(data, keywordList, tu_dien_luong):
    result = {}
    for keyword in keywordList:
        if keyword not in result:
            result[keyword] = {}
        for i in range(len(data)):
            people = f'Người {i + 1}'
            result[keyword][people] = 'ít'
            for sentence in data[i]['sentences']:
                if keyword in sentence:
                    done = False
                    for k in tu_dien_luong.keys():
                        for v in tu_dien_luong[k]:
                            if v in sentence:
                                result[keyword][people] = k
                                done = True
                                break
                        if done:
                            break
                    if not done:
                        result[keyword][people] = 'vừa'
    return result
```
